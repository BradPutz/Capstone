```python
df1 = df1.drop(['POTENTIAL GROSS'],axis = 1)
```

```python
df1.info()
```

```python
df1['AVG TICKET PRICE'] = df1['AVG TICKET PRICE'].str.replace('$','')
df1['AVG TICKET PRICE'] = df1['AVG TICKET PRICE'].astype('float64')
```

```python
df1['TOP TICKET PRICE'] = df1['TOP TICKET PRICE'].str.replace('$','').str.replace(',','')
df1['TOP TICKET PRICE'] = df1['TOP TICKET PRICE'].astype('float64')
```

```python
df1['THIS WEEK GROSS'] = df1['THIS WEEK GROSS'].str.replace('$','').str.replace(',','')
df1['THIS WEEK GROSS'] = df1['THIS WEEK GROSS'].astype('float64')
```

```python
df1['SEATS SOLD'] = df1['SEATS SOLD'].str.replace(',','')
df1['SEATS SOLD'] = df1['SEATS SOLD'].astype('int64')
```

```python
df1['SEATS IN THEATER'] = df1['SEATS IN THEATER'].str.replace(',','')
df1['SEATS IN THEATER'] = df1['SEATS IN THEATER'].astype('int64')
```

```python
df1['CAPACITY %'] = df1['CAPACITY %'].str.rstrip('%').astype(float) / 100
```

```python
df1.info()
```

```python
df1.head(5)
```

```python
df1.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(65,20))
sns.barplot(data=df1.groupby(["THEATER"]).sum("AVG TICKET PRICE"),x="THEATER",y="AVG TICKET PRICE")
```

```python
plt.figure(figsize=(75,20))
sns.barplot(data=df1.groupby(["THEATER"]).sum("AVG TICKET PRICE").reset_index().sort_values(by="AVG TICKET PRICE",
            ascending=False),
            x="THEATER",y="AVG TICKET PRICE")
```

```python
plt.figure(figsize=(15,7))
plt.scatter(data=df1[df1['SHOW'] == 'Hamilton'],x=df1['SHOW'].dt.year,y='THIS WEEK GROSS', color='purple')
plt.title('Hamilton Revenue')
plt.xlabel('Time')
plt.ylabel('Revenue')
plt.show()
```

```python
plt.figure(figsize=(15,7))
plt.scatter(data=df1[df1['SHOW'] == 'The Lion King'],x='WEEK DATE',y='THIS WEEK GROSS', color='red')
plt.title('The Lion King Revenue')
plt.xlabel('Time')
plt.ylabel('Revenue')
plt.show()
```

```python
plt.figure(figsize=(15,7))
plt.scatter(data=df1[df1['SHOW'] == 'Wicked'],x='WEEK DATE',y='THIS WEEK GROSS', color='green')
plt.title('Wicked Revenue')
plt.xlabel('Time')
plt.ylabel('Revenue')
plt.show()
```

```python
plt.figure(figsize=(15,7))
plt.scatter(data=df1[df1['SHOW'] == 'The Phantom of the Opera'],x='WEEK DATE',y='THIS WEEK GROSS', color='black')
plt.title('The Phantom of the Opera Revenue')
plt.xlabel('Time')
plt.ylabel('Revenue')
plt.show()
```

```python
Highest_Revenue = df1.groupby('SHOW')['THIS WEEK GROSS'].sum().reset_index().sort_values(
    by='THIS WEEK GROSS', ascending=False)
```

```python
Top_Ten = Highest_Revenue.head(10)
plt.figure(figsize=(8,5))
sns.barplot(data=Top_Ten,x='THIS WEEK GROSS',y='SHOW')
plt.show()
```

```python
Avg_Ticket = df1.groupby('SHOW')['AVG TICKET PRICE'].mean().reset_index().sort_values(
    by='AVG TICKET PRICE', ascending=False
)
```

```python
plt.figure(figsize=(8,5))
top_10 = Avg_Ticket.head(10)
sns.barplot(data=top_10, x='AVG TICKET PRICE',y='SHOW')
plt.title('Top Shows by Average Ticket Price')
plt.ylabel('Show Title')
plt.xlabel('Average ticket price by week')
plt.show()
```
